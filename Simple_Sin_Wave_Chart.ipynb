```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.arange(0, 10, 0.1)
y_values = np.sin(x_values)

plt.plot(x_values, y_values, label = 'Sin Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Control Sin Chart')

plt.legend()

plt.show()
```
